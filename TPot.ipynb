```python
import tpot
import pandas as pd
from sklearn.model_selection import train_test_split
from tpot import TPOTClassifier
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
df = pd.read_csv('dataset/dataset_39_ecoli_standard.csv')
df.head()
```

```python
Y = df['class'].to_numpy()
X = df[df.columns[:7]].to_numpy()
```

```python
X
```

```python
Y
```

```python
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size=0.3,random_state=22)
```

```python
tp = TPOTClassifier(generations=5,population_size=100,verbosity=2)
```

```python
tp.fit(x_train,y_train)
```

```python
#plot generations 1 - 10, populasi = 100, mutasi_rate = 0.9, crossover_rate = 0.1
x_gen = [i for i in range(1,11)]
hasil = []
for i in x_gen:
    tp = TPOTClassifier(generations=i,population_size=100,verbosity=2)
    tp.fit(x_train,y_train)
    hasil.append(tp.score(x_test,y_test))

plt.plot(x_gen,hasil)
plt.xlabel('Generations')
plt.ylabel('Score')
plt.title('Generations vs Score')
plt.show()
```

```python
#plot generations 4, populasi = 10-200, mutasi_rate = 0.9, crossover_rate = 0.1
x_pop = [i*10 for i in range(1,21)]
hasil = []
for i in x_pop:
    tp = TPOTClassifier(generations=4,population_size=i,verbosity=2)
    tp.fit(x_train,y_train)
    hasil.append(tp.score(x_test,y_test))

```

```python
plt.plot(x_pop,hasil)
plt.xlabel('Populasi')
plt.ylabel('Score')
plt.title('Populasi vs Score')
plt.show()
```

```python
#plot generations 4, populasi = 200, mutasi_rate = 0.1 - 1, crossover_rate = 0.1
x_mut = [i/10 for i in range(1,10)]
hasil = []
for i in x_mut:
    tp = TPOTClassifier(generations=4,population_size=200,mutation_rate=i,verbosity=2)
    tp.fit(x_train,y_train)
    hasil.append(tp.score(x_test,y_test))
```

```python
plt.plot(x_mut,hasil)
plt.xlabel('mutasi_rate')
plt.ylabel('Score')
plt.title('mutasi rate vs Score')
plt.show()
```

```python
#plot generations 4, populasi = 200, mutasi_rate = 0.6, crossover_rate = 0.1 - 1
x_cros = [i/10 for i in range(1,10)]
hasil = []
for i in x_cros:
    tp = TPOTClassifier(generations=4,population_size=200,mutation_rate=0.6,crossover_rate=i,verbosity=2)
    tp.fit(x_train,y_train)
    hasil.append(tp.score(x_test,y_test))
```
