```python
import tpot
import pandas as pd
from sklearn.model_selection import train_test_split
from tpot import TPOTClassifier
```

```python
dataku = pd.read_csv('dataset_39_ecoli.csv')
dataku.head()
```

```python
dictio_cnvrt_class = {"b'cp'":0, "b'im'":1, "b'pp'":2, "b'imU'":3, "b'om'":4, "b'omL'":5, "b'imL'":6, "b'imS'":7}
dataku['class'] = dataku['class'].map(dictio_cnvrt_class)
```

```python
Y = dataku['class'].to_numpy()
X = dataku[dataku.columns[:7]].to_numpy()
```

```python
X
```

```python
Y
```

```python
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size=0.3,random_state=22)
```

```python
x_train
```

```python
x_test
```

```python
y_test
```

```python
tp = TPOTClassifier(generations=5,population_size=100,verbosity=2)
```

```python
tp.fit(x_train,y_train)
```

```python
tp.score(x_test,y_test)
```

```python
tp.export('tpot_ecoli_pipeline.py')
```
